# Method for Generating Graphs with Hamiltonian Cycles

```python
import os
import sys
import random
import numpy as np
import networkx as nx
```

### Distribution 1: Valid Cycle
make_n_cycle: Creates a graph with n nodes and add edges that create a simple loop. Randomly generate additional edges (the number of edges is uniformly sampled from [n, n(n-1)/2] (simple loop to complete connectivity) and add them randomly.

```python
def make_n_cycle(n = 10):
    # Create the cycle graph
    g = nx.generators.classic.cycle_graph(n)
    # Sample the number of additional edges from uniform
    add_num = np.random.randint(n, n*(n-1)/2) - n
    i = 0
    while i < add_num:
        # Generate two nodes to connect
        (a, b) = np.random.choice(range(n), 2, replace=False)
        # Make sure the node doesn't exist yet
        if not g.has_edge(a, b):
            g.add_edge(a, b)
            i = i + 1
    return g
```

```python
g = make_n_cycle(12)
```

```python
nx.draw(g, with_labels = True)
```

make_cycle_dist: Creates list of graphs containing k graphs for each of the node numbers specified in node_nums.

```python
def make_cycle_dist(K, node_nums):
    g_list = []
    for i in range(len(node_nums)):
        n = node_nums[i]
        for k in range(K):
            g = make_n_cycle(n)
            g_list.append(g)
    return g_list
```

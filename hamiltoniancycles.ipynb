# Method for Generating Graphs with Hamiltonian Cycles

```python
import os
import sys
import random
import numpy as np
import networkx as nx
```

### Distribution 1: Valid Cycle
make_n_cycle: Creates a graph with n nodes and add edges that create a simple loop. Randomly generate additional edges (the number of edges is uniformly sampled from [n, n(n-1)/2] (simple loop to complete connectivity) and add them randomly.

```python
def make_n_cycle(n = 10):
    # Create the cycle graph
    g = nx.generators.classic.cycle_graph(n)
    # Sample the number of additional edges from uniform
    add_num = np.random.randint(n, n*(n-1)/2) - n
    i = 0
    while i < add_num:
        # Generate two nodes to connect
        (a, b) = np.random.choice(range(n), 2, replace=False)
        # Make sure the node doesn't exist yet
        if not g.has_edge(a, b):
            g.add_edge(a, b)
            i = i + 1
    return g
```

```python
g = make_n_cycle(12)
```

```python
nx.draw(g, with_labels = True)
```

make_cycle_dist: Creates list of graphs containing k graphs for each of the node numbers specified in node_nums.

```python
def make_cycle_dist(K, node_nums):
    g_list = []
    for i in range(len(node_nums)):
        n = node_nums[i]
        for k in range(K):
            g = make_n_cycle(n)
            g_list.append(g)
    return g_list
```

### Distribution 1: Invalid Cycle


Detecting a Hamiltonian Cycle is NP-complete, so there is no way to efficiently generate a representative distribution of non-Hamiltonian graphs. We will generate some instances by creating a random graph, taking its largest component, and adding additional nodes to it with exactly one edge. Notice that we are excluding unconnected graphs, since those will obviously not have a Hamiltonian cycle and connectivity can be verified easily with BFS.

We will augment these instances with the curated datasets from the [Flinders Hamiltonian Cycle Project](https://sites.flinders.edu.au/flinders-hamiltonian-cycle-project/graph-database/a).

make_n_noncycle: Creates a random graph with n-1 nodes, gets the largest connected component, and adds one node and vertex to connect it to the original graph. Edge probability can also be given, otherwise it defaults to 0.5.

```python
def make_n_noncyle(n, edge_prob = 0.3):
    # Generate the graph with n-1 nodes
    g = nx.generators.random_graphs.gnp_random_graph(n=n - 1, p=edge_prob, seed=None, directed=False)
    # Get the largest component
    components = sorted(nx.connected_components(g), key=len, reverse=True)
    largest_component = components[0]
    # Set and reindex g to be just the largest connected component
    g = nx.relabel.convert_node_labels_to_integers(g.subgraph(largest_component))
    # Add in additional nodes to ensure that it does not have a cycle
    for i in range(len(largest_component), n):
        # Generate a node to connect
        g.add_node(i)
        # Generate a random existing node to connect it to
        a = np.random.randint(0, i)
        # Add an edge between
        g.add_edge(a, i)
    return g
```

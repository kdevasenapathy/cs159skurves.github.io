```python
from networkx.generators.random_graphs import gnp_random_graph
import numpy as np
import networkx as nx


def gen_n_random_graphs(n=10, nodes_range=(10, 50), edge_prob_range=(0.15, 0.65), seed=None):
    graphs = [0]*n
    num_nodes = np.random.randint(low=nodes_range[0], high=nodes_range[1], size=n)
    start_prob, end_prob = edge_prob_range
    edge_prob = np.random.random(size=n)*(end_prob-start_prob)+start_prob
    for i in range(n):
        graphs[i] = gnp_random_graph(n=num_nodes[i], p=edge_prob[i], seed=None, directed=False)
    return graphs
```

# Maximal Independent Set
An independent set is a set of vertices in a graph such that no two of which are adjacent. (sometimes called vertex packing).

Using the algorithm from [Boppana and Halldorsson (1992)](https://networkx.github.io/documentation/latest/reference/algorithms/generated/networkx.algorithms.approximation.independent_set.maximum_independent_set.html?highlight=independent%20set)  here to find the maximal independent set of some random graph G.

```python
from networkx.algorithms.approximation import independent_set
# takes in a undirected graph and returns a aproximate maximal set 
undirected_graphs = gen_n_random_graphs()
```

```python
# Check graph
nx.draw(undirected_graphs[0], with_labels = True)
```

```python
# Find the maximal sets
maximal_sets = []
for g in undirected_graphs:
    # returns the random maximal independent set
    maximal_sets.append(independent_set.maximum_independent_set(g))
```
